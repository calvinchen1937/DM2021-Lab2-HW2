For the preprocessing, I write a clean_text function. First, I lower all the capital characters. Also, I check whether the sentense exists the stopwords and delete them. Then I use re.sub to clean the sentence. I also found that "<lh>" exists at the bottom of each sentences, so I replace it with a "" to delete it.  

````python
def clean_text(text):
    text = text.lower() 
    text = ' '.join(x for x in text.split() if "#" not in x)
    text = ' '.join(x for x in text.split() if "@" not in x)
    text = ' '.join(word for word in text.split() if word not in stopwords.words('english'))
    text = re.sub('[/(){}\[\]\|@,;].',' ', text)
    text = re.sub('[0-9]','', text)
    text = re.sub(r'[^\w\s]','',text)
    text = text.replace('lh', '')
    
    return text
````

Before training the models, I checked the data whether there is nan value. Then I apply tokenizer on the dataset for training, also I change the classes into the different labels.

````python
clean_train = pd.read_csv("clean_train.csv")
clean_test = pd.read_csv("clean_test.csv")

y_try = clean_train.emotion


max_words = 1000
tokenize = Tokenizer(num_words=max_words, char_level=False)

clean_train["text"] = clean_train["text"].fillna("")
clean_test["text"] = clean_test["text"].fillna("")


tokenize.fit_on_texts(clean_train["text"])

x_train = tokenize.texts_to_matrix(clean_train["text"])
x_test = tokenize.texts_to_matrix(clean_test["text"])

encoder = LabelEncoder()
encoder.fit(y_try)
y_train = encoder.transform(y_try)

num_classes = np.max(y_train) + 1
y_train = to_categorical(y_train, num_classes)
````

I first trained a model with DNN models and save the best model for testing. 

````python
batch_size = 100
epochs = 20

model = Sequential()
model.add(Dense(512, input_shape=(max_words,)))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(128))
model.add(Activation('relu'))
model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

filepath = "model/try1.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')

history = model.fit(x_train, y_train, batch_size=batch_size, epochs=epochs, callbacks=checkpoint, verbose=1, validation_split=0.2)
print('training finish')
````
